```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#BARCHART DATA -- CHANNEL 1

#CARE
care_ssim_pred_mean = 0.751355296611922
#UPDATE THIS NUMBER
care_mse_pred_mean = 0.03751162330779352

#PIX2Pix channel 1, sample 500
p2p_ssim_pred_mean = 0.941461369068227
#UPDATE THIS NUMBER
p2p_mse_pred_mean = 0.009427612878409143

#UNET
unet_ssim_pred_mean = 0.5831881658009966
unet_mse_pred_mean = 0.04760294572818723

SSIM_original_mean = 0.3489214022884552, 0.3489214022884552, 0.3489214022884552
```

```python
#SSIM

n_groups = 3

objects = ('CARE Model', 'Pix2Pix', 'UNet')
y_pos = np.arange(len(objects))
performance = [care_ssim_pred_mean,p2p_ssim_pred_mean,unet_ssim_pred_mean]

    
fig, ax = plt.subplots()
ax.bar(y_pos, performance, align='center', alpha=0.5, color='g', label='Mean Predicted SSIM')
ax.plot(y_pos, SSIM_original_mean, "k--", label='Training SSIM')

plt.xticks(y_pos, objects)
plt.ylabel('SSIM')
plt.title('Comparing Model SSIM for Channel 1')

plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.3))


plt.show()
```
